# Trajectories in xy plane for a mesh of data point (x, y)

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def map(B, k):

    xn, yn = B
    
    y_new = np.mod((yn + k*np.sin(xn)), 2*np.pi)
    x_new = np.mod((xn + y_new), 2*np.pi)

    return (x_new, y_new)

# Define k, k is a dimensionless parameter that influences the degree of chaos.
k = 1.21

# Number of iterations
n = 200

# Define the initial x and y
x_o = np.linspace(0, 2*np.pi, 9)
y_o = np.linspace(0, 2*np.pi, 10)

# Create pairs of the initial x and y, this creates points on the xy plane
mesh = [(x, y) for x in x_o for y in y_o]

plt.figure(figsize=(7, 5))

# Define a list where the evolution of each (x, y) pair will be stored
trajectories = []

for point in mesh:

    # Store the initial values of a (x, y) pair
    trajectories = [point]
    
    for i in range(n): # Do n iterations for each pair

        # The new values of (x, y)
        X, Y = map(trajectories[i], k)

        # Store the new values to the list
        trajectories.append((X, Y)) # Along this list the evolution of a (x, y) pair has been imprinted

    # List to array in order to have x and y in columns for easier plotting
    trajectories = np.array(trajectories)

    # Plot the evolution of the (x, y) pair, x coordinates are in the first column, y coordinates are in the second column
    plt.plot(trajectories[:, 0], trajectories[:, 1], '.', markersize=2)
    
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 2 * np.pi)
plt.ylim(0, 2 * np.pi)
plt.title(f"K = {k:.2f}")
plt.show()
```

# Initial condition (xo, yo) = (3.5, 3.5), k = 1.21

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def map(B, k):

    xn, yn = B
    
    y_new = np.mod((yn + k*np.sin(xn)), 2*np.pi)
    x_new = np.mod((xn + y_new), 2*np.pi)

    return (x_new, y_new)

# Define k, k is a dimensionless parameter that influences the degree of chaos.
k = 1.21

# Number of iterations
n = 200

# Create a pair of the initial x and y, this creates the initial data point on the xy plane
mesh = [(3.5, 3.5)]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Store the initial values of a (x, y) pair
trajectories = [mesh[0]]

for i in range(n): # Do n iterations for each pair

    # The new values of (x, y)
    X, Y = map(trajectories[i], k)

    # Store the new values to the list
    trajectories.append((X, Y)) # Along this list the evolution of a (x, y) pair has been imprinted

# List to array in order to have x and y in columns for easier plotting
trajectories = np.array(trajectories)

# Plot the evolution of the (x, y) pair, x coordinates are in the first column, y coordinates are in the second column
ax1.plot(trajectories[:, 0], trajectories[:, 1], '.', markersize=2)
    
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim(0, 2 * np.pi)
ax1.set_ylim(0, 2 * np.pi)
ax1.set_title(f"K = {k:.2f}")
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
ax1.tick_params(axis='both', which='major', labelsize=12)
ax1.tick_params(axis='both', which='minor', labelsize=10)

t = np.linspace(0, 200, len(trajectories[:, 0]))
ax2.plot(t, trajectories[:, 0], label = 'x', color = 'black', linestyle = '--')
ax2.plot(t, trajectories[:, 1], label = 'y', color = "red")
ax2.set_title("Time evolution of x and y")
ax2.legend(loc = "upper right")
ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
ax2.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='minor', labelsize=10)

plt.tight_layout()
plt.show()
```
